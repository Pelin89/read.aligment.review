```python
# percentage of unique mapped reads
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
tool_df1 = tool_df[['sample', 'tool', 'version', 'abs_r']]

g_df = tool_df[['sample', 'tool', 'g:mapped']]
g_df.insert(2, 'version' , 'g')

tool_df1.rename(columns={'abs_r': 'mapped'}, inplace=True)
g_df.rename(columns={'g:mapped': 'mapped'}, inplace=True)
g_df = g_df.drop_duplicates(keep='first')

temp1_df = pd.concat([tool_df1, g_df])
fig1_df = temp1_df.sort_values(by=['tool', 'sample', 'version'], ascending=True)
fig1_df = fig1_df.reset_index(drop=True)


fig1_df['mapped'][fig1_df['sample'] == samp_009390] = fig1_df['mapped']*100/read_009309
fig1_df['mapped'][fig1_df['sample'] == samp_181410] = fig1_df['mapped']*100/read_181410
fig1_df['mapped'][fig1_df['sample'] == samp_183377] = fig1_df['mapped']*100/read_183377


ax1=sns.boxplot(x="tool", y="mapped", hue="version", data=fig1_df, palette="Set1",linewidth=2)
# sns.swarmplot(x="tool", y="unique mapped reads", hue="read version", data=data_df, palette="Set1",linewidth=2)
plt.legend(bbox_to_anchor=(1,1))
plt.ylim(0, 100)

ax1.set(xlabel='tools', ylabel='mapped read percentage')



# fig1_df
```

```python
# percentage of common reads between original and alternative ordered reads

fig2_df = tool_df[['sample', 'tool', 'version', 'r']]

fig2_df['r'][fig2_df['sample'] == samp_009390] = fig2_df['r']*100/read_009309
fig2_df['r'][fig2_df['sample'] == samp_181410] = fig2_df['r']*100/read_181410
fig2_df['r'][fig2_df['sample'] == samp_183377] = fig2_df['r']*100/read_183377

ax2=sns.boxplot(x="tool", y="r", hue="version", data=fig2_df, palette="Set1",linewidth=2)
# sns.swarmplot(x="tool", y="unique mapped reads", hue="read version", data=data_df, palette="Set1",linewidth=2)
plt.legend(bbox_to_anchor=(1,1))
ax2.set(xlabel='tools', ylabel='common mapped read percentage')
plt.ylim(0, 100)

# fig2_df
```
